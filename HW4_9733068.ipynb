```python
'''
____ Author____
Nazanin Mohammadrezaii-9733068

____ COURSE ____
Digital Image processing

____Homework_____
HW6_question 4

_____Intro_______
img2 is the original image,img1 is a tilted and transformed form of img2,consider we are given img2
and we should apply affine transformations to get to original image(img2) with Landmarks 

'''

import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
%matplotlib inline


################# Read the images ##############
img1 = cv.imread('MRI.jpg',0)
img2 = cv.imread('MRI2.jpg',0)

plt.figure(figsize=(8,8))
plt.suptitle('a review on the images',fontsize="x-large", y=.85)
plt.subplot(1,2,1)
plt.imshow(img1,cmap='gray',vmin=0,vmax=255);plt.axis(True);plt.title('MRI: MRI transformed')
plt.axis(False)

plt.subplot(1,2,2)
plt.imshow(img2,cmap='gray',vmin=0,vmax=255);plt.axis(True);plt.title('MRI2: original MRI')
plt.axis(False)
plt.tight_layout()
plt.show()
```

```python
###################### Choose the landmarks #####################

# from img1
landmark1 = np.float32(([164,38],[160,175],[108,63]))  

# from img2
landmark2 = np.float32(([27,102],[255,255],[9,222]))


########## Find the transformation matrix and apply it #########

# Calculate the transformation matrix T to go from landmark1 to landmark2
T= cv.getAffineTransform(landmark1,landmark2)

# Apply the affine transformation on img1
rows,cols = img1.shape
out = cv.warpAffine(img1, T , (cols,rows))

# plot
plt.figure(figsize=(8,8))
plt.imshow(out, cmap = 'gray', vmin = 0, vmax = 255)
plt.title('reconstructed image with the help of landmarks',fontsize="x-large")
plt.axis(False)
plt.tight_layout()
plt.show()
```
